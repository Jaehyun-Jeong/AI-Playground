```python
from pandas import read_csv
import matplotlib.pyplot as plt
from KNN import KNN
```

```python
df = read_csv("../Datasets/zebra_giraffe_crowd_noise_index.csv")
X = df[['x', 'y']].to_numpy()
Y = df['species'].to_numpy()

fig, ax = plt.subplots()
colorMap = {0: 'black', 1: 'orange', 2: 'hotpink', 3: 'green'}
```

# 예시1: 노이즈가 있는 k = 1 분류 (노이즈는 나무, 꽃)

```python
ax.grid(True, which='both')
ax.scatter(X[Y == 0, 0], X[Y == 0, 1],
            s=80,
            color=colorMap[0],
            marker='o')
ax.scatter(X[Y == 1, 0], X[Y == 1, 1],
            s=80,
            color=colorMap[1],
            marker='o')
ax.scatter(X[Y == 2, 0], X[Y == 2, 1],
            s=80,
            color=colorMap[2],
            marker='o')
ax.scatter(X[Y == 3, 0], X[Y == 3, 1],
            s=80,
            color=colorMap[3],
            marker='o')
fig
```

```python
model = KNN(X, Y, k=1, p=1)  # p=1 은 p1-norm 죽, 맨해튼 거리를 의미한다.
result = model.predict([0, 0])
ax.scatter(0, 0,
            s=320,
            color=colorMap[result],
            marker='*')
fig
```
